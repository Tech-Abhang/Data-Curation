check not null values

```python
heart.notnull().sum()
```

drop row containing na values

```python
heart_drop = heart.dropna(axis=0)
heart_drop.shape
```

drop columns containing na values

```python
heart_drop = heart.dropna(axis=1)
heart_drop.shape
```

drop rows with atleast one na value

```python
heart_drop = heart.dropna(how = "any")
heart_drop.shape
```

drop rows if and only if every element is na

```python
heart_drop = heart.dropna(how = "all")
heart_drop.shape
```

if there are not more than 5 notnull values row will drop.

```python
heart_drop = heart.dropna(thresh = 5)
heart_drop.shape
```

```python
heart_drop = heart.dropna(thresh = 20)
heart_drop.shape
```

```python
heart_drop = heart.dropna(subset = ["Age" , "ID"])
heart_drop.shape
```

```python
heart_drop = heart.dropna(subset = ["Gender" , "ID"])
heart_drop.shape
```

mean of age for new data

```python
meanofa = new_data["Age"].mean()
print(meanofa)
```

add mean of age of new data to clo age in na values.

```python
new_data["Age"].fillna(new_data["Age"].mean())
```

cal salary median

```python
medianofs = new_data["Salary"].median()
print(medianofs)
```

add salary median to na values of salary.

```python
new_data["Salary"].fillna(new_data["Salary"].median())
```

```python
new_data["State"].fillna("Unknown")
```
