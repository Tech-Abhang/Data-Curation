```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```python
url = "https://www.nirfindia.org/Rankings/2024/ResearchRanking.html"
response = requests.get(url)
```

```python
print(response.content)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
table = soup.find('table', attrs={'id':'tbl_overall'})
table_body = table.find('tbody')
print(table_body)
```

```python
data = []  # Initialize an empty list to store the extracted data

# Find all 'tr' (table row) elements inside table_body without going deeper into nested tables
rows = table_body.find_all('tr', recursive=False)

# Loop through each row
for row in rows:
    # Find all 'td' (table data/cell) elements inside the row
    cols = row.find_all('td', recursive=False)
    
    # Extract the first content of each column if it exists
    cleaned_cols = []
    for ele in cols:
        if ele.contents:  # Check if the cell has content
            cleaned_cols.append(ele.contents[0])
    
    # Add the cleaned row to data
    data.append(cleaned_cols)
```

```python
rank_data = pd.DataFrame(data)
rank_data.columns= ['Institute ID', 'Name', 'City', 'State', 'Score', 'Rank']
rank_data.to_csv("2024rank_data_research1.csv")
```
