```python
import pandas as pd
import numpy as np

data= pd.read_csv("heart.csv")
data.head()
```

# Normalize caffine intake column

## min-max


```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
caffine_normalized = scaler.fit_transform(data[["Caffeine Intake (mg/day)"]])

print(data.head())
```

```python
min_value = data["Caffeine Intake (mg/day)"].min()
max_value = data["Caffeine Intake (mg/day)"].max()
print(max_value)


data["min_max"] = (data["Age"] - min_value) / (max_value - min_value)
print(data)

```

## z-score


```python
sd = data["Age"].std()
mean = data["Age"].mean()

data["Z-score"] = (data["Age"] - mean) / sd
print(data)
```

## Decimal Scaling

```python
import math
max_abs_value = data["Caffeine Intake (mg/day)"].abs().max()
j = math.ceil(math.log10(max_abs_value))

for i in data["Caffeine Intake (mg/day)"]:
  normalized = i / 10**j
  print(normalized)
```

## Robust Scaling


introduce outlier

```python
data.loc[3,"Caffeine Intake (mg/day)"] = 100000000000000

```

robust scaling

```python
median = data["Caffeine Intake (mg/day)"].median()
Q1 = data["Caffeine Intake (mg/day)"].quantile(0.25)
Q3 = data["Caffeine Intake (mg/day)"].quantile(0.75)
IQR = Q3 - Q1

for i in data["Caffeine Intake (mg/day)"]:
  normalized = (i - median) / IQR
  print(normalized)
```
